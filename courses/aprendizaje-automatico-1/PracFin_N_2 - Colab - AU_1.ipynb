**Contexto:** Se proporcionan las etiquetas correctas (`Real_Label`), los scores predichos por un clasificador (`Predicted_Score`) y la etiqueta predicha (`Predicted_label`). Debemos calcular métricas de evaluación y graficar la curva ROC.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report,
    roc_curve, auc, ConfusionMatrixDisplay
)

# Cargar datos
df = pd.read_csv('predicciones_scores.data')
# Columnas: y_test (Real_Label), y_scores (Predicted_Score), y_pred (Predicted_label)
y_real = df['y_test']
y_scores = df['y_scores']
y_pred = df['y_pred']

print(f"Shape: {df.shape}")
print(f"Columnas: {df.columns.tolist()}")
df.head(10)
```

---
## Pregunta 21 — Curva ROC
Usando los scores proporcionados, graficar la curva ROC y seleccionar la opción correcta.

```python
# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_real, y_scores)
roc_auc = auc(fpr, tpr)

# Graficar
plt.figure(figsize=(7, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', label='Azar (AUC = 0.50)')
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.show()

print(f"\nAUC = {roc_auc:.4f}")
print("\n>>> AUC \u2248 0.83 \u2192 Corresponde al Clasificador 1 \u2192 Respuesta: c")
```

---
## Preguntas 22, 23, 24 — Métricas: Accuracy, Precision, Recall
Calcular las métricas del clasificador usando la matriz de confusión.

```python
# Matriz de confusión
cm = confusion_matrix(y_real, y_pred)
tn, fp, fn, tp = cm.ravel()

print("=" * 40)
print("MATRIZ DE CONFUSI\u00d3N")
print("=" * 40)
print(f"              Pred=0   Pred=1")
print(f"Real=0 ({tn+fp})  TN={tn}   FP={fp}")
print(f"Real=1 ({fn+tp})  FN={fn}    TP={tp}")
print()

# Visualización
fig, ax = plt.subplots(figsize=(6, 5))
ConfusionMatrixDisplay.from_predictions(y_real, y_pred, cmap='Blues', ax=ax)
plt.title('Matriz de Confusi\u00f3n')
plt.show()
```

```python
# Métricas
acc = accuracy_score(y_real, y_pred)
prec = precision_score(y_real, y_pred)
rec = recall_score(y_real, y_pred)
f1 = f1_score(y_real, y_pred)

print("=" * 40)
print("M\u00c9TRICAS DEL CLASIFICADOR")
print("=" * 40)
print(f"P22 \u2014 Accuracy:  {acc:.4f} = {acc*100:.0f}%  \u2192 Respuesta: a (80%)")
print(f"P23 \u2014 Precision: {prec:.4f} = {prec*100:.0f}%  \u2192 Respuesta: b (89%)")
print(f"P24 \u2014 Recall:    {rec:.4f} = {rec*100:.0f}%  \u2192 Respuesta: a (63%)")
print(f"     F1-Score:  {f1:.4f} = {f1*100:.0f}%")
print()
print("Classification Report completo:")
print(classification_report(y_real, y_pred))
```

## Pregunta 25 — Falsos Negativos vs Falsos Positivos
**Verdadero o Falso**: El número de falsos negativos es mayor que el de falsos positivos.

```python
print(f"Falsos Negativos (FN): {fn}")
print(f"Falsos Positivos (FP): {fp}")
print(f"\nFN ({fn}) > FP ({fp})? {fn > fp}")
print("\n>>> Respuesta: VERDADERO")
print("\nEl modelo falla m\u00e1s en detectar positivos reales (los clasifica como negativos)")
print("que en generar falsas alarmas.")
```

## Pregunta 26 — Especificidad > 85%
**Verdadero o Falso**: El clasificador identificó correctamente más de 85% de las instancias de la clase negativa.

```python
specificity = tn / (tn + fp)

print(f"Especificidad = TN / (TN + FP) = {tn} / ({tn} + {fp}) = {specificity:.4f}")
print(f"Especificidad = {specificity*100:.1f}%")
print(f"\n{specificity*100:.1f}% > 85%? {specificity > 0.85}")
print("\n>>> Respuesta: VERDADERO")
print("\nEl clasificador identifica correctamente el 94% de las instancias negativas.")
```
