###  Regresión Logística con Python y Scikit-Learn

### Conceptos clave

### 📌 1️⃣ ¿Qué modela realmente la regresión logística?

En regresión logística no modelamos directamente la probabilidad ( p ).

Modelamos el **logaritmo natural de los momios (odds)**:

$
\log\left(\frac{p}{1-p}\right) = \theta_0 + \theta_1 X_1 + \theta_2 X_2 + \dots + \theta_k X_k
$

donde:

* $ p = P(Y=1 \mid X) $
* $ \frac{p}{1-p} $ son los **momios**
* El logaritmo de los momios se llama **log-momios** o **log-odds**
* $ \theta_i $ son los coeficientes del modelo

---

### 📌 2️⃣ ¿Qué son los momios?

Los **momios** se definen como:

$
\text{momios} = \frac{p}{1-p}
$

Interpretación:

* Si  $p = 0.80 $:

$
\frac{0.80}{0.20} = 4
$

Significa:

> Los momios son 4 a 1 a favor de que ocurra el evento.

* Si $ p = 0.20 $:

$
\frac{0.20}{0.80} = 0.25
$

Significa:

> Los momios son 1 a 4 en contra del evento.

---

### 📌 3️⃣ ¿Qué representa un coeficiente $ \theta_i $?

En el modelo:

$
\log\left(\frac{p}{1-p}\right)=\theta_0 + \theta_1 X_1 + \dots
$

El coeficiente $ \theta_i $ representa:

> El cambio en el **logaritmo de los momios** cuando $ X_i $ aumenta una unidad, manteniendo las demás variables constantes.

Pero el log-momio no es intuitivo.
Por eso exponenciamos el coeficiente.

---

### 📌 4️⃣ Interpretación correcta: $ e^{\theta_i} $

Si aumentamos $ X_i $ en una unidad:

$
\text{Nuevos momios}=e^{\theta_i} \times \text{Momios anteriores}
$

👉 **$e^{\theta_i} $ es un factor multiplicativo sobre los momios.**

---

### 📌 5️⃣ Ejemplo con coeficiente positivo

Supongamos:

$
\theta_1 = 0.7
$

Calculamos:

$
e^{0.7} \approx 2.01
$

Interpretación:

> Por cada unidad adicional de $ X_1 $, los momios se multiplican por 2.01.

Es decir:

* Los momios aproximadamente se duplican.
* La probabilidad aumenta.
* Pero el aumento en probabilidad no es lineal.

---

### 📌 6️⃣ Ejemplo con coeficiente negativo

Supongamos:

$
\theta_1 = -0.43
$

Calculamos:

$
e^{-0.43} \approx 0.65
$

Interpretación:

> Por cada unidad adicional de $ X_1 $, los momios se multiplican por 0.65.

Eso significa:

* Los momios disminuyen en un 35%
$
  1 - 0.65 = 0.35
 $

👉 El evento se vuelve menos probable.

---

### 📌 7️⃣ Ejemplo aplicado (riesgo crediticio)

Supón que:

* Variable: ingresos estandarizados
* $ \theta_{\text{ingresos}} = -0.43 $

Entonces:

$
e^{-0.43} = 0.65
$

Interpretación clara:

> Por cada aumento de una desviación estándar en ingresos, los momios de incumplimiento se reducen en aproximadamente 35%.

No decimos:

❌ “La probabilidad baja 0.43”
❌ “La probabilidad baja 43%”

Decimos:

✅ “Los momios se multiplican por 0.65”

---

### 📌 8️⃣ ¿Por qué no podemos interpretar directamente en probabilidad?

Porque la relación entre log-momios y probabilidad es no lineal:

$
p = \frac{1}{1 + e^{-(\theta_0 + \theta X)}}
$

Un mismo cambio en momios produce distintos cambios en probabilidad dependiendo del punto inicial.

Ejemplo:

Si $ p = 0.10 $:

* Momios = 0.11
* Si se duplican → probabilidad ≈ 0.18

Si $ p = 0.70 $:

* Momios = 2.33
* Si se duplican → probabilidad ≈ 0.82

👉 El efecto en probabilidad depende del punto de partida.

Por eso el modelo es lineal en log-momios, no en probabilidad.

## 📘Regresión Logística – Social Network Ads

El **Social Network Ads Purchase Prediction Dataset** es un conjunto de datos diseñado para el estudio de problemas de **clasificación binaria**, particularmente en el contexto de **Regresión Logística**.

## 📊 Descripción general del dataset

Este dataset contiene información demográfica a nivel de usuario obtenida a partir de una campaña publicitaria en una red social. El objetivo principal es **predecir si un usuario comprará un producto después de haber visto un anuncio**, utilizando como variables predictoras su edad y su salario estimado.

Cada fila del archivo `Social_Network_Ads.csv` representa un usuario individual que fue expuesto a un anuncio publicitario.

El conjunto de datos es estructurado y limpio, lo que significa que requiere un preprocesamiento mínimo y resulta ideal para fines educativos y experimentales en aprendizaje automático supervisado.

---

## 🧾 Variables incluidas

El dataset contiene dos variables independientes (predictoras) y una variable dependiente (objetivo):

### 1️⃣ Age (Edad)

Variable numérica continua. Edad del usuario medida en años. Permite analizar cómo la edad influye en la probabilidad de compra.

### 2️⃣ EstimatedSalary (Salario Estimado)

Variable numérica continua. Ingreso anual estimado del usuario, expresado en dólares. Representa la capacidad económica del usuario, lo que puede influir en su decisión de compra.

### 3️⃣ Purchased (Compra)

* **Tipo:** Variable binaria (0/1)
* **Descripción:**
  * 0 = El usuario no compró el producto
  * 1 = El usuario compró el producto
* **Rol en el modelo:** Variable dependiente (target).
* **Interpretación:** Es la variable que se desea predecir.

### 1️⃣ Introducción y planteamiento formal del problema

Dataset: **Social Network Ads Purchase Prediction Dataset**

### Variables

* $ x_1 = \text{Age} $
* $ x_2 = \text{EstimatedSalary} $
* $ y = \text{Purchased} \in {0,1} $

Queremos modelar:

$
P(y=1 \mid x; \theta)
$

Usando el modelo de regresión logística:

$
h_\theta(x) = \sigma(z)
$

donde:

$
z = \theta_0 + \theta_1 x_1 + \theta_2 x_2
$

y la función sigmoide:

$
\sigma(z) = \frac{1}{1 + e^{-z}}
$

## 5️⃣ Implementación del modelo (Scikit-Learn)

Aunque los estudiantes conocen la teoría, aquí formalizamos la relación con (\theta).

## 6️⃣ Recuperación explícita de los parámetros (\theta)

En sklearn:

* `model.intercept_` → $ \theta_0 $
* `model.coef_` → $ \theta_1, \theta_2 $

El modelo estimado es:

$
\hat{P}(y=1 \mid x) =
\sigma(\theta_0 + \theta_1 x_1 + \theta_2 x_2)
$

donde $x_1$ y $x_2$ están normalizados.



````python
theta_0 = model.intercept_[0]
theta_1 = model.coef_[0][0]
theta_2 = model.coef_[0][1]

print("θ0:", theta_0)
print("θ1 (Age):", theta_1)
print("θ2 (Salary):", theta_2)
````

La frontera ocurre cuando:

$
\theta_0 + \theta_1 x_1 + \theta_2 x_2 = 0
$

Es una recta en el espacio de variables normalizadas.

Se puede graficar para visualizar separabilidad lineal.


````python
# ==============================
# Crear malla para la frontera
# ==============================
x1_min, x1_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
x2_min, x2_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1

xx1, xx2 = np.meshgrid(
    np.arange(x1_min, x1_max, 0.01),
    np.arange(x2_min, x2_max, 0.01)
)

# Predicción sobre la malla
grid = np.array([xx1.ravel(), xx2.ravel()]).T
Z = model.predict(grid)
Z = Z.reshape(xx1.shape)

# ==============================
# Graficar frontera
# ==============================
plt.figure(figsize=(10, 6))

# Frontera de decisión
plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=plt.cm.RdYlBu)

# ==============================
# Graficar datos de entrenamiento
# ==============================
plt.scatter(
    X_train[y_train == 0, 0],
    X_train[y_train == 0, 1],
    c='blue',
    marker='o',
    label='Train - Clase 0'
)

plt.scatter(
    X_train[y_train == 1, 0],
    X_train[y_train == 1, 1],
    c='red',
    marker='o',
    label='Train - Clase 1'
)

# ==============================
# Graficar datos de prueba
# ==============================
plt.scatter(
    X_test[y_test == 0, 0],
    X_test[y_test == 0, 1],
    c='blue',
    marker='x',
    s=100,
    label='Test - Clase 0'
)

plt.scatter(
    X_test[y_test == 1, 0],
    X_test[y_test == 1, 1],
    c='red',
    marker='x',
    s=100,
    label='Test - Clase 1'
)

# ==============================
# Detalles del gráfico
# ==============================
plt.xlabel("Feature 1: Age")
plt.ylabel("Feature 2: Salary")
plt.title("Frontera de Decisión - Regresión Logística")
plt.legend()
plt.show()

````


## 7️⃣ Interpretación rigurosa de los coeficientes

En regresión logística:

$
\log \left( \frac{P(y=1)}{P(y=0)} \right) = 
\theta_0 + \theta_1 x_1 + \theta_2 x_2
$

Es decir:

$
\theta_j = \frac{\partial}{\partial x_j} \log\left(\text{odds}\right)
$

Interpretación:

* Si $( \theta_1 > 0 $: al aumentar la edad, aumentan los log-odds de compra.
* Si $ \theta_2 > 0 $: mayor salario → mayor probabilidad de compra.

En términos de odds ratio:

$
\exp(\theta_j)
$



````python
np.exp(theta_1), np.exp(theta_2)
````

Interpretación:

* Si $e^{\theta_1} = 1.8 $: por cada incremento de 1 desviación estándar en edad, las odds aumentan 80%.

Esto es mucho más informativo que el coeficiente lineal.
